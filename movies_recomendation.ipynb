```python
import ast
```

```python
# CREATING A HELPER FUNCTION WHICH MAINLLY  STORE THE SOME VALUE FROM DICTIONARY AND PUT INTO THE LIST

def convert(object):      # DEFINING A CONVERTER WHICH WILL MANLY CONVERT  A DICTONARY INTO  THE LIST  AS WE CAN RUN THE LOOP ON LIST
  L=[]    # CREATING  A LIST
  count=0
  for i in ast.literal_eval(object):
    if count!=3:
      L.append(i['name'])        # THIS  WILL MAINLY ADD ALL THE NAME RELATED VALUE IN THE LIST L
    else:
      break
    return L         # THIS WILL RETURN THE LIST L
```

```python
data['genres']=data['genres'].apply(convert)
```

```python
data['keywords']=data['keywords'].apply(convert)
```

```python
#

def convert2(obj):
  m=[]                   # CREATING A EMPTY LIST
  counter = 0
  for i in ast.literal_eval(obj):
    if counter !=3:     # only upto three value will be appended
      m.append(i['name'])
      counter +=1    # increasing the counter
    else :
      break
    return m
```

```python
# CREATING A FUNCTION WHOSE NAME IS FETCH_DIRECTOR

def fetch_director(obj):
  n=[]
  for i in ast.literal_eval(obj):
    if i ['job']=='Director':
      n.append(i['name'])
      break
  return n
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```
